```python


def foo(n):
    print(f"Inside foo: n = {n}")
    def bar(x):
        print(f"Inside bar: x = {x}, n = {n}")
        return x + n
    print(f"Returning bar from foo: n = {n}")
    return bar

add_five = foo(5)
result = add_five(10)
print(result)  # This will print 15
```
